```python
import re

def extrair_vendas(texto):
    # Procura números seguidos de 'mil' ou qualquer quantidade com 'mil'
    match = re.search(r'(\d+)(mil)', texto)
    if match:
        return int(match.group(1))  # Retorna apenas o número
    else:
        # Caso não encontre 'mil', tenta pegar o valor numérico sem mil
        match = re.search(r'(\d+)', texto)
        return int(match.group(1)) if match else 0  # Retorna o número, se encontrado

# Aplicar a função na coluna 'venda' para extrair a quantidade
df['quantidade_vendida'] = df['Vendas'].apply(extrair_vendas)
```

```python
df.drop(['quantidade_vendida'], axis = 1)
```

```python
def ajustar_quantidade(texto):
    if '5mil' in texto:
        return 5000
    elif '10mil' in texto:
        return 10000
    elif '50mil' in texto:
        return 50000
    elif '250mil' in texto:
        return 250000
    elif '100mil' in texto:
        return 100000
    else:
        # Caso o texto não contenha "mil", extrair o número diretamente
        return int(''.join(filter(str.isdigit, texto))) if any(c.isdigit() for c in texto) else 0

# Aplicar a função na coluna 'venda'
df['quantidade_vendida'] = df['Vendas'].apply(ajustar_quantidade)
```
