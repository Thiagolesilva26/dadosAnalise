```python
df['Caracter Padronizado'] = (df['Caracter Descricao'] - df['Caracter Descricao'].mean()) / df['Caracter Descricao'].std()
```

```python
df['Imagem Padronizado'] = (df['Quantidade Imagem'] - df['Quantidade Imagem'].mean()) / df['Quantidade Imagem'].std()
```

```python
df['quantidade padronizada'] = (df['quantidade_vendida'] - df['quantidade_vendida'].mean()) / df['quantidade_vendida'].std()
```

```python
df.drop(['Quantidade Imagem'], axis = 1, inplace=True)
```

```python
df.drop(['Caracter Descricao'], axis = 1, inplace=True)
```

```python
df.drop(['quantidade_vendida'], axis = 1, inplace=True)
```
